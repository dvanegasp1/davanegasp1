```python
# prompt: cree una columna donde se asigne la categoria de la columna "GRADO_CARGO" segun corresponda teniendo en cunata la siguiente logica:
# ST = 1
# SI = 1
# TE = 2
# IT = 2
# CT = 3
# IJ = 3
# MY = 4
# SC = 4
# TC = 5
# CM = 5
# CR = 6

import pandas as pd

# Assuming 'datos' DataFrame is already loaded as shown in the prompt
# datos = pd.read_excel('/content/trABAJO ULTIMO DE ANALISIS TARDE 24102024.xlsx', sheet_name='todo')

def categorize_grade(grade):
  if grade in ['ST', 'SI']:
    return 1
  elif grade in ['TE', 'IT']:
    return 2
  elif grade in ['CT', 'IJ']:
    return 3
  elif grade in ['MY', 'SC']:
    return 4
  elif grade in ['TC', 'CM']:
    return 5
  elif grade == 'CR':
    return 6
  else:
    return 0 # or handle other cases as needed


datos['CATEGORIA_CARGO'] = datos['GRADO_CARGO'].apply(categorize_grade)
datos
```

```python
# prompt: ordanice los registros de la taba "datos", con referencia a la 'IDENTIFICACION', 'CATEGORIA' y 'JERARQUIA'

datos = datos.sort_values(by=['IDENTIFICACION', 'CATEGORIA_CARGO', 'JERARQUIA'])
datos
```

```python
# prompt: realice un conteo de los datos faltantes de la columna "FECHA_FINAL"

# Count missing values in 'FECHA_FINAL' column
missing_dates = datos['FECHA_FINAL'].isnull().sum()
print(f"NÃºmero de datos faltantes en la columna 'FECHA_FINAL': {missing_dates}")
```

```python
# prompt: VERIFIQUE QUE LA COLUMNA "FECHA FINAL" SEA SUPERIOR A FECHA "ID_FECHFISC_ASCENSO"

# Assuming 'datos' DataFrame is already loaded and processed as shown in the previous code.

def compare_dates(row):
    try:
        fecha_final = pd.to_datetime(row['FECHA_FINAL'], errors='coerce')
        id_fechfisc_ascenso = pd.to_datetime(row['ID_FECHFISC_ASCENSO'], errors='coerce')

        if fecha_final > id_fechfisc_ascenso:
            return True
        else:
            return False
    except (ValueError, TypeError):
        return False  # Handle cases where date conversion fails

datos['FECHA_COMPARACION'] = datos.apply(compare_dates, axis=1)
datos
```

```python
# prompt: identifique las categorias mas altas de la columna "CATEGORIA_CARGO",CON REFERENCIA A LAS  CEDULAS DE LA COLUMNA "identificacion" y elimine las que este por debajo de la mas alta,

import pandas as pd

# Assuming 'datos' DataFrame is already loaded and processed as shown in the previous code.

def categorize_grade(grade):
  if grade in ['ST', 'SI']:
    return 1
  elif grade in ['TE', 'IT']:
    return 2
  elif grade in ['CT', 'IJ']:
    return 3
  elif grade in ['MY', 'SC']:
    return 4
  elif grade in ['TC', 'CM']:
    return 5
  elif grade == 'CR':
    return 6
  else:
    return 0

datos['CATEGORIA_CARGO'] = datos['GRADO_CARGO'].apply(categorize_grade)

# Group by 'IDENTIFICACION' and get the maximum 'CATEGORIA_CARGO' for each ID
max_categories = datos.groupby('IDENTIFICACION')['CATEGORIA_CARGO'].max()

# Merge the maximum categories back into the original DataFrame
datos = datos.merge(max_categories.rename('MAX_CATEGORIA'), left_on='IDENTIFICACION', right_index=True)

# Filter out rows where 'CATEGORIA_CARGO' is less than 'MAX_CATEGORIA'
datos_filtered = datos[datos['CATEGORIA_CARGO'] == datos['MAX_CATEGORIA']]

# Drop the 'MAX_CATEGORIA' column (optional)
datos_filtered = datos_filtered.drop('MAX_CATEGORIA', axis=1)

datos = datos_filtered
datos
```
